## Nomor 7

Sebuah perusahaan asuransi menggunakan
periode seleksi 2 tahun. Diketahui:

• 𝑞[50] = 0,0010 (mortalita
pertama seleksi)

• 𝑞[50]+1 = 0,0015 (mortalita
kedua seleksi)

• 𝑞52 =0,0025 (mortalita ultimate pada
usia 52)

Hitunglah:

a. 𝑝[50]
2 (probabilitas seseorang yang terpilih
pada usia 50 akan hidup 2 tahun)

b. 𝑝52 (probabilitas seseorang usia 52 akan
hidup setahun)

### a. 𝑝[50] 2 (probabilitas seseorang yang terpilih pada usia 50 akan hidup 2 tahun)

#### manual

image.png

#### phyton

```python
# a. 𝑝[50] 2 (probabilitas seseorang yang terpilih pada usia 50 akan hidup 2 tahun)

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data
q_50_sel1 = 0.0010
q_50_sel2 = 0.0015

# Probabilitas hidup
p_50_sel1 = 1 - q_50_sel1
p_50_sel2 = 1 - q_50_sel2
p_50_2 = p_50_sel1 * p_50_sel2

print("Probabilitas seseorang yang terpilih pada usia 50 akan hidup 2 tahun:", round(p_50_2,6))

# Data usia
usia_a = np.array([50, 51, 52])
survival_a = [1, p_50_sel1, p_50_2]

sns.set_style("whitegrid")

plt.figure(figsize=(8,5))
sns.lineplot(x=usia_a, y=survival_a, marker='o')

plt.title("Soal 7a: Kurva Survival Seleksi Usia 50")
plt.xlabel("Usia")
plt.ylabel("Probabilitas")
plt.ylim(0.996, 1.001)

for x, y in zip(usia_a, survival_a):
    plt.text(x, y+0.0001, f"{y:.6f}", ha='center')

plt.tight_layout()
plt.show()
```

```python
# b. 𝑝52 (probabilitas seseorang usia 52 akan hidup setahun)

# Data
q_52 = 0.0025
p_52 = 1 - q_52

print("Probabilitas hidup usia 52 selama 1 tahun:", round(p_52,6))

# Data usia
usia_b = np.array([52, 53])
survival_b = [1, p_52]

plt.figure(figsize=(8,5))
sns.lineplot(x=usia_b, y=survival_b, marker='o')

plt.title("Soal 7b: Kurva Survival Ultimate Usia 52")
plt.xlabel("Usia")
plt.ylabel("Probabilitas")
plt.ylim(0.996, 1.001)

for x, y in zip(usia_b, survival_b):
    plt.text(x, y+0.0001, f"{y:.6f}", ha='center')

plt.tight_layout()
plt.show()
```

## Nomor 8

Dua orang saudara kembar berusia tepat 40
tahun. Dengan asumsi mortalita mereka
independen dan menggunakan tabel mortalita
dengan 𝑝40 = 0,998 dan 𝑝41 = 0,9979,
hitunglah probabilitas bahwa keduanya akan
hidup mencapai usia 42 tahun.

### manual

WhatsApp Image 2026-03-05 at 8.54.44 PM.jpeg

### phyton

```python
# NOMOR 8

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data
p40 = 0.998
p41 = 0.9979

# Perhitungan
p40_2 = p40 * p41
prob_keduanya = p40_2 ** 2

print("2p_40 =", round(p40_2, 6))
print("Probabilitas keduanya hidup =", round(prob_keduanya, 6))

# Data waktu
t_values = np.array([40, 41, 42])

survival_1 = [1, p40, p40_2]
survival_2 = [s**2 for s in survival_1]

# DataFrame
df_survival = pd.DataFrame({
    't': t_values,
    'Survival_1_Orang': survival_1,
    'Survival_2_Orang': survival_2
})

sns.set_style('whitegrid')

fig, axes = plt.subplots(2, 1, figsize=(9, 10))

# ===============================
# Kurva Survival 1 Orang
# ===============================
sns.lineplot(ax=axes[0], x='t', y='Survival_1_Orang',
             data=df_survival, marker='o')

axes[0].set_title('Kurva Survival 1 Orang (Usia 40 → 42)')
axes[0].set_xlabel('usia')
axes[0].set_ylabel('Probabilitas')
axes[0].set_ylim(0.98, 1.01)

# Tambahkan angka di titik
for x, y in zip(t_values, survival_1):
    axes[0].text(x, y+0.0003, f"{y:.6f}", ha='center')

# ===============================
# Kurva Survival 2 Orang
# ===============================
sns.lineplot(ax=axes[1], x='t', y='Survival_2_Orang',
             data=df_survival, marker='o')

axes[1].set_title('Kurva Survival 2 Orang Independen')
axes[1].set_xlabel('usia')
axes[1].set_ylabel('Probabilitas')
axes[1].set_ylim(0.98, 1.01)

# Tambahkan angka di titik
for x, y in zip(t_values, survival_2):
    axes[1].text(x, y+0.0003, f"{y:.6f}", ha='center')

plt.tight_layout()
plt.show()
```
