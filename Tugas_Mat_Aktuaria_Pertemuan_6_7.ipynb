## NOmor 6

Manual

WhatsApp Image 2026-03-12 at 7.07.56 PM.jpeg

Phyton

## Nomor 12
. Sebuah perusahaan asuransi menggunakan
periode seleksi 2 tahun. Diketahui:
l[60] = 10.000

q[60] = 0,004

q[60]+1 = 0,005

q62 = 0,007 (ultimate)
Hitunglah:
a. 𝑙[60]+1 dan 𝑙[60]+2

b. Probabilitas seseorang yang terpilih pada

usia 60 akan hidup hingga usia 63 (𝑝[60]
3 )

Manual

WhatsApp Image 2026-03-12 at 7.07.06 PM.jpeg WhatsApp Image 2026-03-12 at 7.07.38 PM.jpeg

Phyton

```python
l60 = 10000

q60 = 0.004
q60_1 = 0.005
q62 = 0.007

# jumlah hidup
l60_1 = l60 * (1 - q60)
l60_2 = l60_1 * (1 - q60_1)

# probabilitas hidup 3 tahun
p3_60 = (1 - q60) * (1 - q60_1) * (1 - q62)

print("l[60]+1 =", l60_1)
print("l[60]+2 =", l60_2)
print("3p[60] =", p3_60)
```
